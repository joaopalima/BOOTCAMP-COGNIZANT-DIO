```python
import pandas as pd
import pandera as pa
```

```python
#dados foram extraídos
# 'na_values=valores_ausentes' vai susbstituir os dados de valores_ausentes por dados não disponíveis(NaN)

valores_ausentes = ['**','*****','****','####','###!','NULL']
df = pd.read_csv('ocorrencia.csv', sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
df.head(10) #imprime as primeiras 10 linhas
```

```python
#etapa de validaação dos dados
schema = pa.DataFrameSchema(
    columns = {
        'codigo_ocorrencia': pa.Column(pa.Int),
        'codigo_ocorrencia2': pa.Column(pa.Int),
        'ocorrencia_classificacao': pa.Column(pa.String),
        'ocorrencia_cidade': pa.Column(pa.String),
        'ocorrencia_uf': pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True), ##retirando os ascterísticos na coluna estado
        'ocorrencia_aerodromo': pa.Column(pa.String, nullable=True),
        'ocorrencia_dia': pa.Column(pa.DateTime),
        'ocorrencia_hora': pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True), ##adicionando valores nulos na coluna// pertindo valor nulo
        'total_recomendacoes': pa.Column(pa.Int)
    }
)
```

```python
schema.validate(df)
## mostrou data frama, foi validado e possui dados do tipo inteiro
```

```python
#combinar 3 filtros
# ocorrencia cuja (classificação == INCIDENTE GRAVE ou lassificação == INCIDENTE) e o estado == SP
# verifica se a classificação está dentro de >>>> classificacao.isin(['INCIDENTE GRAVE','INCIDENTE'])

filtro1 = df.ocorrencia_classificacao.isin(['INCIDENTE GRAVE','INCIDENTE'])
filtro2 = df.ocorrencia_uf == 'SP'
#df.loc[filtro] #imprimes todos os valores
df.loc[filtro & filtro2] # comparador lógico ou (|) traz os dois dados
```

```python
#ocorrencias cjuja cidade começa com a letra "c"

filtro = df.ocorrencia_cidade.str[0] =='C' #índice do primeiro caracter '0' letra "c"

#imprimes todos os valores
df.loc[filtro]
```

```python
#ocorrencias cuja cidade terminam com a letra 'A'

filtro = df.ocorrencia_cidade.str[-1] =='A' #índice do últimao caracter '-1' letra "A"

#imprimes todos os valores de cidade que terminam com o caracter "A"
df.loc[filtro]
```

```python
#ocorrencias cuja cidade terminam com os caracteres 'MA'

filtro = df.ocorrencia_cidade.str[-2:] =='MA' #índice do últimao caracter '-2:'(penultimo em diante) letra "MA"

#imprimes todos os valores de cidade que terminam com o caracter "MA"
df.loc[filtro]
```

```python
#ocorrencias cuja cidade contém os caracteres 'MA'

filtro = df.ocorrencia_cidade.str.contains('MA') #índice do verrifica se contém letras "MA" // qualquer parte do texto no conteúdo da coluna

#imprimes todos os valores de cidade que os caracteres "MA" em qualquer parte do conteúdo
df.loc[filtro]
```

```python
#ocorrencias cuja cidade contém os caracteres 'MA' ou 'AL'

filtro = df.ocorrencia_cidade.str.contains('MA | AL') #índice do verrifica se contém letras "MA" ou "AL" // qualquer parte do texto no conteúdo da coluna

#imprimes todos os valores de cidade que os caracteres "MA" ou 'AL' em qualquer parte do conteúdo
df.loc[filtro]
```

```python
#filtors nas datas
#obter ocorrencias do ano de 2015

filtro = df.ocorrencia_dia.dt.year == 2015 ## extrai o ano através do "dt.year" com valor == 2015

#imprimes todos os valores de ano de 2015
df.loc[filtro]

```

```python
#filtors nas datas
#obter ocorrencias do ano de 2015 e mês 12

filtro1 = df.ocorrencia_dia.dt.year == 2015 ## extrai o no através do "dt.year" com valor == 2015
filtro2 = df.ocorrencia_dia.dt.month == 12  ## extrai o no através do "dt.month" com valor == 12

#imprimes todos os valores de ano de 2015
df.loc[filtro1 & filtro2]
```

```python
#filtors nas datas
#obter ocorrencias do ano de 2015 e mês 12

## extrai o no através do "dt.year" com valor == 2015 e extrai o no através do "dt.month" com valor == 12
filtro = (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12)

#imprimes todos os valores de ano de 2015
df.loc[filtro]
```

```python
#filtors nas datas
#obter ocorrencias do ano de 2015 e mês 12 e pegar o dia 8

## extrai o no através do "dt.year" com valor == 2015 e extrai o no através do "dt.month" com valor == 12 e dia == 8(dt.day == 8)
filtro = (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12) & (df.ocorrencia_dia.dt.day == 8)

#imprimes todos os valores de ano de 2015
df.loc[filtro]
```

```python
#filtors nas datas
#obter ocorrencias do ano de 2015 e mês 12 e pegar o dia 3 ao dia 08

## extrai o no através do "dt.year" com valor == 2015 e extrai o no através do "dt.month" com valor == 12 e dia >=3 & <8
filtro = (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12) & ((df.ocorrencia_dia.dt.day >=3) & (df.ocorrencia_dia.dt.day <+ 8))

#imprimes todos os valores de ano de 2015
df.loc[filtro]
```

```python
# criar nova coluna para filtrar dia e hora 
#transofrar datetime em string, pois hora já é uma string
# transforar o dia como se fosse uma string -- astype(str)
#atribuir a nova coluna com pd.to_datetime
df['ocorrencia_dia_hora'] = pd.to_datetime(df.ocorrencia_dia.astype(str) + ' ' + df.ocorrencia_hora)
```

```python
#filtors nas datas de forma mais detahalda

filtro_ano = df.ocorrencia_dia_hora.dt.year == 2015 ## extrai o no através do "dt.year" com valor == 2015
filtro_mes = df.ocorrencia_dia_hora.dt.month == 12  ## extrai o no através do "dt.month" com valor == 12
filtro_dia_inicio = df.ocorrencia_dia_hora.dt.day > 2 ## extrai o no através do "dt.day > 2
filtro_dia_fim = df.ocorrencia_dia_hora.dt.day < 9  ## extrai o no através do "dt.day < 9


#imprimes todos os valores de ano de 2015
df.loc[filtro_ano & filtro_mes & filtro_dia_inicio & filtro_dia_fim]
```
